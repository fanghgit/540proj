```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 16})
```

```python
#rcv1
data_newton = np.loadtxt("rcv1_newton.csv", delimiter = ',')
data_grad = np.loadtxt("rcv1_prox-grad.csv", delimiter = ',')
data_sgd = np.loadtxt("rcv1_sgd.csv", delimiter = ',')
data_newton = data_newton[1:data_newton.shape[0],:]
data_grad = data_grad[1:data_grad.shape[0],:]
data_sgd = data_sgd[1:data_sgd.shape[0],:]
```

```python
plt.figure()
#plt.plot(data[:,0], 1 - data[:,2], 'b', label = "BPR-L2-Naive")
#plt.plot(data_bns[:,0], 1 - data_bns[:,2], 'r', label = "BPR-L2-Balanced")
opt = 5483
plt.plot(  data_newton[:,0], (data_newton[:,1]-opt)/opt, 'b', label="prox-newton" )
plt.plot(  data_grad[:,0], (data_grad[:,1]-opt)/opt, 'r', label="prox-gradient" )
plt.plot( data_sgd[:,0], (data_sgd[:,1]-opt)/opt, 'g', label="prox-sgd" )
plt.xlabel("time(sec)")
plt.yscale("log")
plt.ylabel("relative error (log scale)")
plt.title("Reuters")
#plt.legend(prop={'size': 16})
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.legend(prop={'size': 12},loc=1)
plt.savefig("rcv1.png",bbox_inches='tight')
#plt.show()
```
